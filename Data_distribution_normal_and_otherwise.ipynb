```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
distribution_random = np.random.normal(0, 1, 100)
plt.hist(distribution_random,bins = 20, color = 'c')
plt.axvline(distribution_random.mean(),color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(distribution_random.mean() + distribution_random.std(),color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(distribution_random.mean() - distribution_random.std(),color = 'b', linestyle = 'dashed', linewidth = 2)
plt.title ('Histogram of normal distribution')
plt.show()
```

```python
distribution_poisson = np.random.poisson(4,100)
plt.hist(distribution_poisson,bins = 20, color = 'c')
plt.axvline(distribution_poisson.mean(),color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(distribution_poisson.mean() + distribution_poisson.std(),color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(distribution_poisson.mean() - distribution_poisson.std(),color = 'b', linestyle = 'dashed', linewidth = 2)
plt.title ('Histogram of poisson distribution')
plt.show()
```

```python
distribution_gamma = np.random.gamma(5,2,100)
plt.hist(distribution_gamma,bins = 20, color = 'c')
plt.axvline(distribution_gamma.mean(),color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(distribution_gamma.mean() + distribution_gamma.std(),color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(distribution_gamma.mean() - distribution_gamma.std(),color = 'b', linestyle = 'dashed', linewidth = 2)
plt.title ('Histogram of gamma distribution')
plt.show()
```

```python
distribution_binomial = np.random.binomial(10, 0.6, 100)
plt.hist(distribution_binomial,bins = 20,color = 'c')
plt.axvline(distribution_binomial.mean(),color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(distribution_binomial.mean() + distribution_binomial.std(),color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(distribution_binomial.mean() - distribution_binomial.std(),color = 'b', linestyle = 'dashed', linewidth = 2)
plt.title ('Histogram of binomial distribution')
plt.show()
```

```python
distribution_bernoulli = np.random.binomial(1, 0.6,100)
plt.hist(distribution_bernoulli,bins = 20,color = 'c')
plt.axvline(distribution_bernoulli.mean(),color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(distribution_bernoulli.mean() + distribution_bernoulli.std(),color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(distribution_bernoulli.mean() - distribution_bernoulli.std(),color = 'b', linestyle = 'dashed', linewidth = 2)
plt.title ('Histogram of bernoulli distribution')
plt.show()
```

```python
distribution_standard_t = np.random.standard_t(10, size=100)
plt.hist(distribution_standard_t,bins = 20, color = 'c')
plt.axvline(distribution_standard_t.mean(),color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(distribution_standard_t.mean() + distribution_standard_t.std(),color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(distribution_standard_t.mean() - distribution_standard_t.std(),color = 'b', linestyle = 'dashed', linewidth = 2)
plt.title ('Histogram of bernoulli distribution')
plt.show()
```

Among above 6 types of distributions,  bernoulli is not a good case to use mean and standard deviation to discribe. When the variable value of certain distribution can randomly equal to any number around the mean as center, the distribution is suitable to use mean and standard deviation to describe.

Additionally:

Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1.
Add them together to create a third variable.
Graph the third variable using a histogram.
Compute the mean and standard deviation and plot them as vertical lines on the histogram.
Evaluate the descriptive statistics against the data.

```python
nor1 = np.random.normal(5,0.5,1000)
nor2 = np.random.normal(10,1,1000)
nor3 = nor1 +nor2
```

```python
plt.hist(nor3,bins = 20, color = 'c')
plt.axvline(nor3.mean(),color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(nor3.mean() + nor3.std(),color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(nor3.mean() - nor3.std(),color = 'b', linestyle = 'dashed', linewidth = 2)
plt.title ('Histogram of the third variables distribution')
plt.show()
```

The mean of third variables equals the sum of the first two, however, the standard deviation is between the first two.
